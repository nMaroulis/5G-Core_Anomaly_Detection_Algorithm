### <font color=coral>1. Data Preprocessing and Normalization</font>

```python
def transform_dataset(data):
    #keep only the test UE with IMSI 1 and service type 1
    data = data[(data['imsi'] == 1) & (data['service_type'] == 1)]
    # Certain
    data = data.drop(columns = ['time','x','y','imsi','velocity','z', 'Simulation_Scenario', 'x.1', 'y.1', 'z.1', 'max_battery', 'battery(mAh)',
                                        'battery_status', 'device_status', 'charging_status', 'cellid', 'resource_blocks',
                                        'packet_intervalUL(ms)','sum_ultx_ue(bytes)', 'sum_ulrx_cell(packets)',
                                        'sum_ultx_ue(packets)', 'sum_dlrx_ue(packets)', 'sum_dlrx_ue(bytes)', 'current_data',
                                        'service_type', 'sum_ulrx_cell(bytes)', '#ues', 'txpower(dBm)', 'rsrq', 'cell_type',
                                        'packet_sizeUL(bytes)', 'sum_dltx_cell(packets)', 'sum_dltx_cell(bytes)',
                                        'packet_intervalDL(ms)', 'ultx_ue(bytes)','packet_sizeDL(bytes)','ultx_ue(packets)', 'dlrx_ue(packets)', 
                                        'dlrx_ue(bytes)','dltx_cell(packets)','ulrx_cell(packets)', 'dltx_cell(bytes)'])

    data.loc[data['lost_packets'] < 0, 'lost_packets'] = 0
    data = data.reset_index(drop=True)
    
    return data
```

```python
train_data = data_simulation_2.copy()
test_data = data_simulation_1.copy()

train_data = transform_dataset(train_data)
test_data = transform_dataset(test_data)

#-------------------------------
test_data_UPM = test_data.copy()
test_data_UPM = test_data_UPM.rename(columns = {'ulrx_cell(bytes)': 'ulrx_cell', 'associated_cellid': 'cell_id'}, inplace = False)
test_data_UPM['timestamp'] = 'yyyy-MM-dd-HH'
test_data_UPM = test_data_UPM[['timestamp','ul_delay','dl_delay','lost_packets','rsrp','transfer_protocol','ulrx_cell','cell_id']]
test_data_UPM.to_csv('../4. UPM Dataset/ad_dataset.csv',index = False) # export data for UPM
#_------------------------------

print('The Size of the ' + colored('Train','green') +  ' Set is: ' +  str(train_data.shape))
print('The Size of the ' + colored('Test','red') +  ' Set is: ' +  str(test_data.shape))

print("The "  + colored ('trainining','green') +  " data set has the following features:")
for col in train_data.columns:
    print("\u2022 "+col)

feature_size = len(train_data.columns)

""" Transform Data Normalization/Standardization"""

scaler = MinMaxScaler()
train_data = pd.DataFrame(scaler.fit_transform(train_data), columns=train_data.columns)
test_data = pd.DataFrame(scaler.transform(test_data), columns=test_data.columns)

# save the scaler
joblib.dump(scaler, "../models/scaler.save") 
```

```python
def get_unique_numbers(numbers):

    list_of_unique_numbers = []
    unique_numbers = set(numbers)
    for number in unique_numbers:
        list_of_unique_numbers.append(number)
    return list_of_unique_numbers
```

> ### <font color=Coral>4.1 Deep Neural Network Training Phase </font>

```python
def build_and_run_model(X, y,path_of_model):
    EPOCHS = 150
    VALIDATION_SPLIT = 0.25
    BATCH_SIZE = 32
    VERBOSE = 2

    keras_model = Sequential()
    keras_model.add(Dense(units=64, input_dim=feature_size,activation='relu'))
    keras_model.add(Dense(units=64, input_dim=1,activation='relu'))
    keras_model.add(Dense(units=32, input_dim=1,activation='relu'))
    keras_model.add(Dense(num_of_generated_clusters, activation='softmax'))
    keras_model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
    es = EarlyStopping(monitor='val_loss', mode='min',patience = 7)

    history = keras_model.fit(X, y, validation_split=VALIDATION_SPLIT, epochs=EPOCHS, batch_size=BATCH_SIZE, 
                                 verbose=VERBOSE, shuffle=True,callbacks=[es])
    
    keras_model.save(path_of_model+'model.h5')
    with open(path_of_model+'trainHistory', 'wb') as file_pi:
        pickle.dump(history.history, file_pi)
    print("Saved model and history to disk")
    
    return keras_model
```

```python
""" Fit Data for Model"""
X = train_data_labeled.iloc[:, 0:feature_size].astype(float).values
y = train_data_labeled['Cluster'].values
path_of_model = "../models/"

""" One Hot Encode Label """
encoder = LabelEncoder()
encoder.fit(y)
encoded_y = encoder.transform(y)
y = tf.keras.utils.to_categorical(encoded_y)


keras_model = build_and_run_model(X,y,path_of_model)
```

```python
history = pickle.load(open(path_of_model+'trainHistory', "rb"))
 
fig,ax = plt.subplots(1,2,figsize=(15,5))
ax[0].minorticks_on()
ax[0].grid(which='minor', linestyle=':', linewidth='0.5', color='black')

ax[0].plot(history['accuracy'], label='Accuracy (training data)')
ax[0].plot(history['val_accuracy'], label='Accuracy (validation data)')
ax[0].set_ylabel('Accuracy (%)',fontsize = 14)
ax[0].set_xlabel('Number of Epochs',fontsize = 14)
ax[0].legend(loc="best")

ax[1].minorticks_on()
ax[1].grid(which='minor', linestyle=':', linewidth='0.5', color='black')
ax[1].plot(history['loss'], label='Loss (training data)')
ax[1].plot(history['val_loss'], label='Loss (validation data)')
# ax.set_title('Mean Squared Error for Forecasting Future Behaviour for QoS metrics',fontsize = 18)
ax[1].set_ylabel('Loss',fontsize = 14)
ax[1].set_xlabel('Number of Epochs',fontsize = 14)
ax[1].legend(loc="best")

fig.savefig('../3. Output/Validation.png', bbox_inches='tight')
```
